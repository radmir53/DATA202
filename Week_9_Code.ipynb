```r
# Probability
```

```r
x <- 1 : 10
px <- rep(1 / length(x), length(x))
```

### Scatterplots and linearity

```r
barplot(px, names.arg = x, space = 0, ylim = c(0, 1), xlab = expression(X), ylab = expression(p(x)))
```

```r
# Mean and variance
EX <- sum(x * px)
VX <- mean((x - EX)^ 2)

c(EX, VX)
```

```r
# Random draws from distribution
n <- 100
draws <- runif(n)
```

```r
plot(draws, ylim = c(0, 1), pch = 16, col = alpha("black", 0.5), cex = 2)
```
