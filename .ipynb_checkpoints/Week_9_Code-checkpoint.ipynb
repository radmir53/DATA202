```r
# Set random number generator seed for replicability of results.
set.seed(0)

# Define the variables x and y.
x <- 1 : 10
y <- rnorm(length(x))
```

```r
model <- lm(y ~ x)
```

### Scatterplots and linearity

```r
plot(x, y, xlab = expression(X), ylab = expression(Y), pch = 16, col = gray(0.5), cex = 2)
abline(model, lwd = 2, col = 2)
```
